```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((512,512),dtype=np.uint8)
```

```python
blank
```

```python
text = cv.putText(blank, 'PKL KEREN', (50,250), cv.FONT_HERSHEY_COMPLEX, 2, (255,255,255), thickness=5, lineType=cv.LINE_AA)
```

```python
def show_img(img):
    fig = plt.figure(figsize=(20,20))
    axes = fig.add_subplot(111)
    axes.imshow(img, cmap='gray')
    
```

```python
show_img(text)
```

```python
kernel = np.ones((5,5), dtype=np.uint8)
```

```python
kernel
```

```python
morph_gradient = cv.morphologyEx(text, cv.MORPH_GRADIENT, kernel)
show_img(morph_gradient)
```
